```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"D:\Data Science\Iris unsupervised data.csv")
```

```python
dataset.head()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range (2,21):
    km = KMeans( n_clusters=i ,init='k-means++')
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10, 5))
plt.plot([i for i in range(2, 21)], wcss, marker="o")  # Use "o" for circle markers
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")  # Label for the y-axis
plt.xticks([i for i in range(2, 21)])  # Tick marks for x-axis
plt.title("WCSS vs Number of Clusters")
plt.grid(True)  # Enable grid lines
plt.show()
```

```python
# from uper graph i will chose 3 no of clusters bcz there is elbow curve at 3
```

### Making the models

```python
kmn = KMeans(n_clusters=3)
dataset["Predict"] = kmn.fit_predict(dataset)
```
